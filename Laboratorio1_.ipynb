## 1 - ETL Extración, Tranformación y Carga

### 1. Extración

````python

from google.colab import drive
import pandas as pd
import chardet #libreria para los encoding

#Mostramos Google Drive para acceder al archivo
drive.mount('/content/drive', force_remount=True)

ruta_archivo = '/content/drive/MyDrive/Analisis de datos 2025/Tendencias del mercado laboral de IA/ai_job_market.csv'

with open(ruta_archivo, 'rb') as f:
  #podemos limitar los kbites f.read(5000)
  result = chardet.detect(f.read())


try:
  df = pd.read_csv(ruta_archivo, encoding=result['encoding'], sep=',', decimal='.')

  print('Extración de datos exitosa')
except FileNotFoundError:
  print('El archivo no se encuentra en la ruta especificada')


````

### 2.   Transformación



````python
#   estructura general
df.info()
print("- OBSERVACIÓN: 11 son de tipo object, lo que indica que contienen texto o cadenas de caracteres.Solo 1 columna (job_id) es de tipo int64, es decir, numérica entera.")
df.head()
df.describe(include='all')


````

 ** Explicación de estadísticas en df.describe()**

- **count:** cantidad de valores no nulos en la columna.
- **unique:** cantidad de valores distintos (*solo aplica a columnas categóricas).*
- **top:** valor más frecuente *(la moda)* en columnas categóricas.
- **freq:** frecuencia del valor más común *(cuántas veces aparece el 'top').*
- **mean:** promedio de los valores numéricos.
- **std:** desviación estándar, mide la dispersión respecto al promedio.
- ** min:** valor mínimo en la columna.
- ** 25%:** primer cuartil, el 25% de los datos están por debajo de este valor.
- **50%:** segundo cuartil o mediana, el valor central de la distribución.
- **75%:** tercer cuartil, el 75% de los datos están por debajo de este valor.
- **max:** valor máximo en la columna.



````python
#Renombrando de columnas
df = df.rename(columns={
    'company_name': 'Empresa',
    'industry': 'Industria',
    'job_title': 'Puesto',
    'skills_required': 'Habilidades',
    'experience_level': 'Experiencia',
    'employment_type': 'Tipo Empleo',
    'location': 'Ubicación',
    'salary_range_usd': 'Rango Salarial',
    'posted_date': 'Fecha Publicación',
    'company_size': 'Tamaño Original',
    'tools_preferred': 'Herramientas',
    'Salario Min': 'Salario_Min',
    'Salario Max': 'Salario_Max',
    'Salario Promedio': 'Salario_Promedio',
    'Año': 'Año_Publicación',
    'Mes': 'Mes_Publicación',
    'Tamaño Empresa': 'Tamaño_Estimado',
    'Cantidad Herramientas': 'Cantidad_Herramientas'
})

# Convercion de fechas
df['Fecha Publicación'] = pd.to_datetime(df['Fecha Publicación'], errors='coerce')

# Convercion de numéricos
df['Salario_Min'] = pd.to_numeric(df['Salario_Min'], errors='coerce')
df['Salario_Max'] = pd.to_numeric(df['Salario_Max'], errors='coerce')
df['Salario_Promedio'] = pd.to_numeric(df['Salario_Promedio'], errors='coerce')
df['Cantidad_Herramientas'] = pd.to_numeric(df['Cantidad_Herramientas'], errors='coerce')

df.info()
````

````python
# Eliminación de duplicados
df = df.drop_duplicates()

#  Ver cantidad de nulos por columna
df.isnull().sum()

#  Eliminación de filas con datos clave faltantes
df = df.dropna(subset=['Puesto', 'Rango Salarial'])

# Rellenado de valores faltantes en columnas categóricas
df['Tamaño_Estimado'] = df['Tamaño_Estimado'].fillna('Desconocido')
df['Experiencia'] = df['Experiencia'].fillna(df['Experiencia'].mode()[0])

#  Rellenado de valores faltantes en columnas numéricas
df['Salario_Promedio'] = df['Salario_Promedio'].fillna(df['Salario_Promedio'].mean())
df['Cantidad_Herramientas'] = df['Cantidad_Herramientas'].fillna(0)

# Estructura final
df.describe(include='all')

````

### 3.   Carga



````python
# Ruta de destino en mi Google Drive
ruta_salida = '/content/drive/MyDrive/Analisis de datos 2025/Tendencias del mercado laboral de IA/ai_job_market_limpio.csv'

# Guardado de DataFrame limpio
df.to_csv(ruta_salida, index=False)

print(" Dataset limpio guardado correctamente.")
````

## 2 - EDA Análisis Exploratorio de datos

````python
df['Puesto'].value_counts().head(10)
df['Industria'].value_counts()
df['Experiencia'].value_counts()

#Entry (nivel inicial) perfiles junior
#Mid y Senior: diversificado en niveles de experiencia.

#¿Qué nivel de experiencia es más demandado?
#Conclusión: Las empresas buscan más perfiles Entry-level,
#lo que indica alta demanda de talento junior en el mercado de IA.
````

````python
df.groupby('Experiencia')['Salario_Promedio'].mean().round(2)

#¿Cómo varía el salario según el nivel?

# Entry-level tiene un salario promedio más alto que Mid, lo cual es inusual.
# Senior sí tiene el salario más alto, como se espera.

# La diferencia entre niveles es muy pequeña, lo que sugiere que:
# Puede haber ofertas junior con salarios inflados (por ubicación, industria o herramientas).
# O que Mid-level está subvalorado en algunas empresas.
````

````python
from collections import Counter

# ¿Qué herramientas son más solicitadas por nivel?

# Función para contar herramientas por nivel
def herramientas_por_nivel(nivel):
    herramientas = df[df['Experiencia'] == nivel]['Herramientas'].dropna().str.split(',')
    flat_list = [h.strip() for sublist in herramientas for h in sublist if isinstance(sublist, list)]
    return Counter(flat_list).most_common(5)

niveles = ['Entry', 'Mid', 'Senior']
for nivel in niveles:
    print(f"\n -> Herramientas más pedidas en nivel {nivel}:")
    for herramienta, cantidad in herramientas_por_nivel(nivel):
        print(f"- {herramienta}: {cantidad} veces")

# Conclusión general
# Herramientas como LangChain, FastAPI y KDB+ son transversales a todos los niveles.
# A medida que aumenta la experiencia, se incorporan herramientas más complejas como MLflow, TensorFlow y Scikit-learn.
# Esto refleja una progresión técnica clara: desde el acceso y despliegue hasta el modelado y gestión avanzada.
````
